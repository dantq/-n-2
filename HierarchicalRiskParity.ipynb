## C√°c k·ªπ thu·∫≠t ph√¢n c·ª•m

üîπ Gi·ªõi thi·ªáu:
-  L√† thu·∫≠t to√°n ph√¢n c·ª•m ph·ªï bi·∫øn nh·∫•t.

üîπ C√°c b∆∞·ªõc:

üîπ ∆Øu ƒëi·ªÉm:

üîπ Nh∆∞·ª£c ƒëi·ªÉm:
- C·∫ßn ch·ªâ ƒë·ªãnh s·ªë c·ª•m k tr∆∞·ªõc.

### Ph√¢n c·ª•m ph√¢n c·∫•p (Hierarchical Clustering) 

üîπ Kh√°i ni·ªám:
- T·∫°o ra c·∫•u tr√∫c ph√¢n c·∫•p gi·ªØa c√°c c·ª•m.
- Kh√¥ng c·∫ßn ch·ªâ ƒë·ªãnh s·ªë c·ª•m tr∆∞·ªõc.

üîπ Hai d·∫°ng ch√≠nh

- Divisive (ph√¢n chia) ‚Äì "t·ª´ tr√™n xu·ªëng": b·∫Øt ƒë·∫ßu t·ª´ c·ª•m l·ªõn, chia d·∫ßn.

üîπ Bi·ªÉu di·ªÖn
-  D√πng Dendrogram ‚Äì bi·ªÉu ƒë·ªì c√¢y th·ªÉ hi·ªán qu√° tr√¨nh g·ªôp nh√≥m.

üîπ ∆Øu ƒëi·ªÉm
-  Kh√¥ng c·∫ßn bi·∫øt tr∆∞·ªõc s·ªë c·ª•m.
-    Hi·ªÉn th·ªã quan h·ªá ph√¢n c·∫•p r√µ r√†ng.
-    D·ªÖ tri·ªÉn khai, tr·ª±c quan.

üîπ Nh∆∞·ª£c ƒëi·ªÉm
-  T·ªën th·ªùi gian t√≠nh to√°n v·ªõi d·ªØ li·ªáu l·ªõn.
-    Nh·∫°y c·∫£m v·ªõi ngo·∫°i lai.
-    Dendrogram d·ªÖ b·ªã r·ªëi n·∫øu qu√° nhi·ªÅu ƒëi·ªÉm.


### Ph√¢n c·ª•m Affinity Propagation

üîπ Gi·ªõi thi·ªáu:
- Kh√¥ng c·∫ßn bi·∫øt tr∆∞·ªõc s·ªë c·ª•m.
- D·ª±a tr√™n vi·ªác truy·ªÅn th√¥ng ƒëi·ªáp gi·ªØa c√°c ƒëi·ªÉm d·ªØ li·ªáu.
- T·ª± ƒë·ªông t√¨m ra "exemplars" ‚Äì ƒëi·ªÉm ƒë·∫°i di·ªán c·ªßa m·ªói c·ª•m.

üîπ C∆° ch·∫ø
- Nh·∫≠n v√†o ma tr·∫≠n ƒë·ªô t∆∞∆°ng ƒë·ªìng gi·ªØa c√°c c·∫∑p ƒëi·ªÉm.
- C·∫≠p nh·∫≠t hai lo·∫°i th√¥ng ƒëi·ªáp:
  - Responsibility: ƒëi·ªÉm X c√≥ ph√π h·ª£p l√†m ƒë·∫°i di·ªán cho Y kh√¥ng?
  - Availability: ƒëi·ªÉm Y c√≥ s·∫µn s√†ng nh·∫≠n X l√†m ƒë·∫°i di·ªán kh√¥ng?
- L·∫∑p l·∫°i qu√° tr√¨nh ƒë·∫øn khi h·ªôi t·ª•.

üîπ Tham s·ªë ch√≠nh
- preference: quy·∫øt ƒë·ªãnh s·ªë c·ª•m ‚Üí c√†ng cao th√¨ c√†ng nhi·ªÅu c·ª•m.
- damping: gi√∫p tr√°nh dao ƒë·ªông khi c·∫≠p nh·∫≠t.

üîπ ∆Øu ƒëi·ªÉm
- Kh√¥ng c·∫ßn bi·∫øt tr∆∞·ªõc s·ªë c·ª•m.
- Hi·ªáu qu·∫£ v·ªõi d·ªØ li·ªáu kh√¥ng tuy·∫øn t√≠nh.
- ·ªîn ƒë·ªãnh h∆°n K-Means do kh√¥ng ph·ª• thu·ªôc kh·ªüi t·∫°o.

üîπ Nh∆∞·ª£c ƒëi·ªÉm
- Chi ph√≠ t√≠nh to√°n cao (O(n¬≤)).
- Ch·ªâ ph√π h·ª£p v·ªõi d·ªØ li·ªáu nh·ªè ho·∫∑c v·ª´a.
- Vi·ªác ch·ªçn gi√° tr·ªã preference kh√° kh√≥ ƒëi·ªÅu ch·ªânh.


## CASE STUDY ‚Äì √ÅP D·ª§NG PH√ÇN C·ª§M PH√ÇN C·∫§P TRONG PH√ÇN B·ªî R·ª¶I RO PH√ÇN C·∫§P (HRP)

### L√†m sao ƒë·ªÉ ph√¢n b·ªï t√†i s·∫£n nh·∫±m c√¢n b·∫±ng gi·ªØa r·ªßi ro v√† l·ª£i nhu·∫≠n, ƒë·∫∑c bi·ªát trong ƒëi·ªÅu ki·ªán th·ªã tr∆∞·ªùng kh√¥ng ·ªïn ƒë·ªãnh?

### L√Ω do l·ª±a ch·ªçn Hierarchical Clustering trong ƒë·∫ßu t∆∞
Trong th·ª±c t·∫ø, c√°c t√†i s·∫£n t√†i ch√≠nh th∆∞·ªùng c√≥ xu h∆∞·ªõng di chuy·ªÉn theo nh√≥m ‚Äì v√≠ d·ª•, c·ªï phi·∫øu ng√†nh ng√¢n h√†ng th∆∞·ªùng bi·∫øn ƒë·ªông c√πng nhau. Vi·ªác x√°c ƒë·ªãnh c√°c nh√≥m t√†i s·∫£n c√≥ t∆∞∆°ng quan cao s·∫Ω gi√∫p nh√† ƒë·∫ßu t∆∞:

- Hi·ªÉu ƒë∆∞·ª£c m·ªëi quan h·ªá ·∫©n gi·ªØa c√°c t√†i s·∫£n.

- Gi·∫£m thi·ªÉu r·ªßi ro h·ªá th·ªëng b·∫±ng c√°ch tr√°nh ƒë·∫ßu t∆∞ d·ªìn v√†o c√°c t√†i s·∫£n c√≥ quan h·ªá m·∫≠t thi·∫øt.

- ƒêa d·∫°ng h√≥a hi·ªáu qu·∫£ h∆°n, nh·ªù v√†o c√°i nh√¨n tr·ª±c quan t·ª´ c√¢y ph√¢n c·∫•p (dendrogram) do thu·∫≠t to√°n ph√¢n c·ª•m t·∫°o ra.

Hierarchical Risk Parity (HRP) l√† m·ªôt ph∆∞∆°ng ph√°p ph√¢n b·ªï t√†i s·∫£n ti√™n ti·∫øn, ƒë∆∞·ª£c ƒë·ªÅ xu·∫•t b·ªüi Marcos L√≥pez de Prado ‚Äì chuy√™n gia n·ªïi ti·∫øng trong lƒ©nh v·ª±c t√†i ch√≠nh ƒë·ªãnh l∆∞·ª£ng.
Kh√°c v·ªõi c√°c ph∆∞∆°ng ph√°p truy·ªÅn th·ªëng nh∆∞ Markowitz‚Äôs Mean-Variance Optimization v·ªën d·ª±a tr√™n vi·ªác ngh·ªãch ƒë·∫£o ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai (m·ªôt b∆∞·ªõc th∆∞·ªùng g√¢y m·∫•t ·ªïn ƒë·ªãnh), HRP s·ª≠ d·ª•ng c·∫•u tr√∫c ph√¢n c·∫•p t·ª± nhi√™n trong d·ªØ li·ªáu ƒë·ªÉ:

- Ph√¢n nh√≥m t√†i s·∫£n d·ª±a tr√™n t∆∞∆°ng quan.

- √Åp d·ª•ng thu·∫≠t to√°n "ph√¢n b·ªï r·ªßi ro theo ph√¢n c·∫•p" thay v√¨ c·ªë g·∫Øng t·ªëi ∆∞u to√†n c·ª•c.

- Gi·∫£m ph·ª• thu·ªôc v√†o c√°c gi·∫£ ƒë·ªãnh th·ªëng k√™ l√Ω t∆∞·ªüng (nh∆∞ ph√¢n ph·ªëi chu·∫©n c·ªßa l·ª£i su·∫•t).

üîç ∆Øu ƒëi·ªÉm ch√≠nh c·ªßa HRP:

- Kh√¥ng y√™u c·∫ßu ∆∞·ªõc l∆∞·ª£ng ch√≠nh x√°c ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai.

- ·ªîn ƒë·ªãnh h∆°n trong ƒëi·ªÅu ki·ªán th·ªã tr∆∞·ªùng bi·∫øn ƒë·ªông.

- Ph√¢n b·ªï v·ªën h·ª£p l√Ω h∆°n gi·ªØa c√°c nh√≥m t√†i s·∫£n.



### Thi·∫øt k·∫ø nghi√™n c·ª©u √°p d·ª•ng HRP
üéØ M·ª•c ti√™u:
√Åp d·ª•ng ph√¢n c·ª•m ph√¢n c·∫•p (Agglomerative Clustering) tr√™n t·∫≠p d·ªØ li·ªáu t√†i s·∫£n t√†i ch√≠nh ƒë·ªÉ x√¢y d·ª±ng danh m·ª•c ƒë·∫ßu t∆∞ d·ª±a theo chi·∫øn l∆∞·ª£c HRP.

üß© Quy tr√¨nh nghi√™n c·ª©u:

| B∆∞·ªõc                                           | M√¥ t·∫£                                                                                                           |
| ---------------------------------------------- | --------------------------------------------------------------------------------------------------------------- |
| **B1. Thu th·∫≠p d·ªØ li·ªáu**                       | L·∫•y d·ªØ li·ªáu l·ª£i su·∫•t h√†ng ng√†y c·ªßa c√°c t√†i s·∫£n (c·ªï phi·∫øu, ETF, v.v.).                                           |
| **B2. T√≠nh to√°n ma tr·∫≠n t∆∞∆°ng quan**           | S·ª≠ d·ª•ng h·ªá s·ªë t∆∞∆°ng quan Pearson ho·∫∑c Spearman ƒë·ªÉ ƒëo m·ª©c ƒë·ªô ƒë·ªìng bi·∫øn.                                          |
| **B3. X√¢y d·ª±ng dendrogram**                    | √Åp d·ª•ng ph√¢n c·ª•m ph√¢n c·∫•p ƒë·ªÉ x√°c ƒë·ªãnh c√°c nh√≥m t√†i s·∫£n t∆∞∆°ng ƒë·ªìng.                                              |
| **B4. S·∫Øp x·∫øp t√†i s·∫£n theo c·∫•u tr√∫c ph√¢n c·∫•p** | Duy·ªát qua c√¢y ph√¢n c·∫•p ƒë·ªÉ thi·∫øt l·∫≠p th·ª© t·ª± ph√¢n b·ªï.                                                             |
| **B5. Ph√¢n b·ªï v·ªën theo HRP**                   | √Åp d·ª•ng nguy√™n t·∫Øc ‚Äúr·ªßi ro ngang b·∫±ng gi·ªØa c√°c nh√°nh‚Äù ƒë·ªÉ chia v·ªën.                                              |
| **B6. ƒê√°nh gi√° hi·ªáu qu·∫£**                      | So s√°nh hi·ªáu su·∫•t danh m·ª•c HRP v·ªõi danh m·ª•c Markowitz truy·ªÅn th·ªëng qua Sharpe Ratio, Volatility, Drawdown, v.v. |

B∆∞·ªõc ƒë·∫ßu ti√™n l√† t√¨m c√°c c·ª•m (cluster) d·ª±a tr√™n m·ªëi t∆∞∆°ng quan b·∫±ng k·ªπ thu·∫≠t ph√¢n c·ª•m ph√¢n c·∫•p k·∫øt d√≠nh (agglomerative hierarchical clustering). L·ªõp hierarchy c√≥ ph∆∞∆°ng th·ª©c dendrogram, ph∆∞∆°ng th·ª©c n√†y nh·∫≠n gi√° tr·ªã tr·∫£ v·ªÅ t·ª´ ph∆∞∆°ng th·ª©c linkage c√πng l·ªõp.

Linkage th·ª±c hi·ªán ph√¢n c·ª•m th·ª±c t·∫ø ch·ªâ trong m·ªôt d√≤ng m√£, v√† tr·∫£ v·ªÅ danh s√°ch c√°c c·ª•m ƒë∆∞·ª£c gh√©p l·∫°i theo ƒë·ªãnh d·∫°ng: Z = [c·ªï phi·∫øu_1, c·ªï phi·∫øu_2, kho·∫£ng c√°ch, s·ªë l∆∞·ª£ng m·∫´u].

```python
def correlDist(corr):
    # Ma tr·∫≠n kho·∫£ng c√°ch d·ª±a tr√™n h·ªá s·ªë t∆∞∆°ng quan, v·ªõi 0 <= d[i,j] <= 1
    # ƒê√¢y l√† m·ªôt metric kho·∫£ng c√°ch h·ª£p l·ªá
    dist = ((1 - corr) / 2.)**0.5  # ma tr·∫≠n kho·∫£ng c√°ch
    return dist
```

```python
# T√≠nh to√°n ma tr·∫≠n kho·∫£ng c√°ch d·ª±a tr√™n ma tr·∫≠n t∆∞∆°ng quan c·ªßa d·ªØ li·ªáu l·ª£i nhu·∫≠n
dist = correlDist(returns.corr())
# Th·ª±c hi·ªán ph√¢n c·ª•m theo ph∆∞∆°ng ph√°p linkage v·ªõi thu·∫≠t to√°n 'ward'
link = linkage(dist, 'ward')
# Xem k·∫øt qu·∫£ c·ª•m ƒë·∫ßu ti√™n trong danh s√°ch k·∫øt qu·∫£ linkage
link[0]
```

C√°c l√° c·ªßa c√¢y ch√≠nh l√† t·ª´ng c·ªï phi·∫øu ri√™ng l·∫ª, c√≤n g·ªëc c√¢y l√† c·ª•m duy nh·∫•t cu·ªëi c√πng. Tr·ª•c tung (y-axis) th·ªÉ hi·ªán ‚Äúkho·∫£ng c√°ch‚Äù gi·ªØa c√°c c·ª•m, do ƒë√≥, c√°c nh√°nh c√†ng d√†i th√¨ hai c·ª•m ƒë√≥ c√†ng √≠t c√≥ s·ª± t∆∞∆°ng quan v·ªõi nhau.

## C√°c b∆∞·ªõc th·ª±c hi·ªán Hierarchical Risk Parity (HRP)


#### Ph√¢n ch√©o g·∫ßn ƒë√∫ng (Quasi-diagonalization) v√† l·∫•y tr·ªçng s·ªë cho Hierarchical Risk Parity (HRP)



Ph√¢n ch√©o g·∫ßn ƒë√∫ng (quasi-diagonalization) l√† m·ªôt qu√° tr√¨nh th∆∞·ªùng ƒë∆∞·ª£c bi·∫øt ƒë·∫øn v·ªõi t√™n g·ªçi s·∫Øp x·∫øp ma tr·∫≠n (matrix seriation) v√† c√≥ th·ªÉ th·ª±c hi·ªán b·∫±ng ph∆∞∆°ng ph√°p ph√¢n c·ª•m ph√¢n c·∫•p (hierarchical clustering).

Qu√° tr√¨nh n√†y t√°i s·∫Øp x·∫øp ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai sao cho c√°c kho·∫£n ƒë·∫ßu t∆∞ t∆∞∆°ng t·ª± ƒë∆∞·ª£c ƒë·∫∑t g·∫ßn nhau. Vi·ªác ph√¢n ch√©o ma tr·∫≠n n√†y gi√∫p ta ph√¢n b·ªï tr·ªçng s·ªë m·ªôt c√°ch t·ªëi ∆∞u theo nguy√™n t·∫Øc ph√¢n b·ªï ngh·ªãch ƒë·∫£o ph∆∞∆°ng sai (inverse-variance allocation).


```python
def getQuasiDiag(link):
    # Chuy·ªÉn d·ªØ li·ªáu link v·ªÅ ki·ªÉu int ƒë·ªÉ d·ªÖ x·ª≠ l√Ω
    link = link.astype(int)
    # Kh·ªüi t·∫°o sortIx l√† m·ªôt Series g·ªìm 2 ph·∫ßn t·ª≠ cu·ªëi c√πng c·ªßa link (c√°c cluster l·ªõn nh·∫•t)
    sortIx = pd.Series([link[-1, 0], link[-1, 1]])
    # L·∫•y s·ªë l∆∞·ª£ng ph·∫ßn t·ª≠ g·ªëc ban ƒë·∫ßu (s·ªë t√†i s·∫£n)
    numItems = link[-1, 3]
    # L·∫∑p cho ƒë·∫øn khi t·∫•t c·∫£ ph·∫ßn t·ª≠ trong sortIx ƒë·ªÅu l√† ch·ªâ s·ªë g·ªëc (kh√¥ng ph·∫£i cluster)
    while sortIx.max() >= numItems:
        # ƒê·ªïi l·∫°i index cho sortIx ƒë·ªÉ ch√®n th√™m c√°c ph·∫ßn t·ª≠ m·ªõi
        sortIx.index = range(0, sortIx.shape[0] * 2, 2)
        # L·∫•y c√°c ph·∫ßn t·ª≠ ƒëang l√† cluster (c√≥ ch·ªâ s·ªë l·ªõn h∆°n ho·∫∑c b·∫±ng numItems)
        df0 = sortIx[sortIx >= numItems]
        i = df0.index            # ch·ªâ s·ªë c·ªßa c√°c cluster trong sortIx
        j = df0.values - numItems # chuy·ªÉn sang ch·ªâ s·ªë trong link (b·∫Øt ƒë·∫ßu t·ª´ 0)
        # Thay th·∫ø cluster ƒë·∫ßu ti√™n b·∫±ng c√°c ph·∫ßn t·ª≠ con cluster ƒë√≥
        sortIx[i] = link[j, 0]   # ph·∫ßn t·ª≠ con 1
        df0 = pd.Series(link[j, 1], index=i + 1)  # ph·∫ßn t·ª≠ con 2, index +1 ƒë·ªÉ ch√®n v√†o v·ªã tr√≠ k·∫ø ti·∫øp
        # Th√™m ph·∫ßn t·ª≠ con th·ª© 2 v√†o sortIx
        sortIx = pd.concat([sortIx, df0]) 
        # S·∫Øp x·∫øp l·∫°i theo index
        sortIx = sortIx.sort_index()
        # ƒê√°nh l·∫°i index li√™n ti·∫øp t·ª´ 0
        sortIx.index = range(sortIx.shape[0])
    # Tr·∫£ v·ªÅ danh s√°ch c√°c ch·ªâ s·ªë t√†i s·∫£n theo th·ª© t·ª± quasi-diagonalized
    return sortIx.tolist()
```

#### Recursive bisection (Ph√¢n chia ƒë·ªá quy)

B∆∞·ªõc n√†y ph√¢n b·ªï tr·ªçng s·ªë th√¥ng qua ph∆∞∆°ng ph√°p ph√¢n chia ƒë·ªá quy (recursive bisection) d·ª±a tr√™n ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai c·ªßa c√°c c·ª•m t√†i s·∫£n.


```python
# H√†m getClusterVar: T√≠nh ph∆∞∆°ng sai c·ªßa m·ªôt c·ª•m t√†i s·∫£n
def getClusterVar(cov, cItems):
    # T√≠nh ph∆∞∆°ng sai cho m·ªôt c·ª•m t√†i s·∫£n (cluster)
    cov_ = cov.loc[cItems, cItems]  # Tr√≠ch xu·∫•t ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai con t∆∞∆°ng ·ª©ng v·ªõi c√°c t√†i s·∫£n trong c·ª•m
    w_ = getIVP(cov_).reshape(-1, 1)  # L·∫•y tr·ªçng s·ªë theo ph∆∞∆°ng ph√°p ngh·ªãch ƒë·∫£o ph∆∞∆°ng sai (Inverse Variance Portfolio)
    cVar = np.dot(np.dot(w_.T, cov_), w_)[0, 0]  # T√≠nh ph∆∞∆°ng sai c√≥ tr·ªçng s·ªë cho c·ª•m ƒë√≥
    return cVar

# H√†m getRecBipart: Ph√¢n b·ªï tr·ªçng s·ªë b·∫±ng ph√¢n ƒë√¥i ƒë·ªá quy (recursive bisection)
def getRecBipart(cov, sortIx):
    # T√≠nh ph√¢n b·ªï tr·ªçng s·ªë HRP (Hierarchical Risk Parity)
    w = pd.Series(1, index=sortIx)  # Kh·ªüi t·∫°o t·∫•t c·∫£ tr·ªçng s·ªë b·∫±ng 1
    cItems = [sortIx]  # T·∫•t c·∫£ t√†i s·∫£n ban ƒë·∫ßu n·∫±m trong m·ªôt c·ª•m
    while len(cItems) > 0:
        # T√°ch m·ªói c·ª•m hi·ªán t·∫°i th√†nh hai c·ª•m con
        cItems = [i[j:k] for i in cItems for j, k in ((0, len(i) // 2), (len(i) // 2, len(i))) if len(i) > 1]
        # Duy·ªát t·ª´ng c·∫∑p c·ª•m
        for i in range(0, len(cItems), 2):
            cItems0 = cItems[i]      # C·ª•m con 1
            cItems1 = cItems[i + 1]  # C·ª•m con 2
            cVar0 = getClusterVar(cov, cItems0)  # Ph∆∞∆°ng sai c·ª•m 1
            cVar1 = getClusterVar(cov, cItems1)  # Ph∆∞∆°ng sai c·ª•m 2

            alpha = 1 - cVar0 / (cVar0 + cVar1)  # T√≠nh t·ªâ l·ªá ph√¢n b·ªï theo ƒë·ªô r·ªßi ro (ph∆∞∆°ng sai)
            w[cItems0] *= alpha      # Tr·ªçng s·ªë c·ª•m 1
            w[cItems1] *= 1 - alpha  # Tr·ªçng s·ªë c·ª•m 2
    return w  # Tr·∫£ v·ªÅ ph√¢n b·ªï cu·ªëi c√πng
```

#### So s√°nh v·ªõi c√°c ph∆∞∆°ng ph√°p ph√¢n b·ªï t√†i s·∫£n kh√°c:

Gi·∫£ ƒë·ªãnh ch√≠nh c·ªßa nghi√™n c·ª©u t√¨nh hu·ªëng n√†y
l√† ph√°t tri·ªÉn m·ªôt ph∆∞∆°ng ph√°p ph√¢n b·ªï t√†i s·∫£n thay th·∫ø cho **Danh m·ª•c t·ªëi thi·ªÉu ph∆∞∆°ng sai (MVP)** c·ªßa Markowitz.

V√¨ v·∫≠y, trong b∆∞·ªõc n√†y, ch√∫ng ta s·∫Ω ƒë·ªãnh nghƒ©a c√°c h√†m ƒë·ªÉ **so s√°nh hi·ªáu su·∫•t** c·ªßa c√°c ph∆∞∆°ng ph√°p ph√¢n b·ªï t√†i s·∫£n sau:

1. **MVP** ‚Äì Danh m·ª•c T·ªëi thi·ªÉu Ph∆∞∆°ng sai c·ªßa Markowitz
2. **HRP** ‚Äì Ph√¢n b·ªï r·ªßi ro theo ph√¢n c·ª•m ph√¢n c·∫•p (Hierarchical Risk Parity)


```python
# H√†m t√≠nh Minimum-Variance Portfolio (MVP):
def getMVP(cov):
    # Chuy·ªÉn ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai sang numpy array
    cov = cov.T.values
    n = len(cov)  # s·ªë l∆∞·ª£ng t√†i s·∫£n
    N = 100
    mus = [10 ** (5.0 * t / N - 1.0) for t in range(N)]  # t·∫≠p h·ª£p tham s·ªë tr∆°n (mu) cho ƒë∆∞·ªùng bi√™n hi·ªáu qu·∫£
    # Chuy·ªÉn sang ma tr·∫≠n c·ªßa th∆∞ vi·ªán cvxopt
    S = opt.matrix(cov)
    pbar = opt.matrix(np.ones(cov.shape[0]))  # gi·∫£ ƒë·ªãnh l·ª£i su·∫•t k·ª≥ v·ªçng ƒë·ªÅu b·∫±ng 1
    # T·∫°o c√°c ma tr·∫≠n r√†ng bu·ªôc
    G = -opt.matrix(np.eye(n))  # r√†ng bu·ªôc tr·ªçng s·ªë >= 0
    h = opt.matrix(0.0, (n, 1))
    A = opt.matrix(1.0, (1, n))  # t·ªïng tr·ªçng s·ªë = 1
    b = opt.matrix(1.0)
    # Gi·∫£i b√†i to√°n t·ªëi ∆∞u h√≥a b·∫≠c hai ƒë·ªÉ t√¨m tr·ªçng s·ªë t·ªëi ∆∞u
    solvers.options['show_progress'] = False
    portfolios = [solvers.qp(mu * S, -pbar, G, h, A, b)['x'] for mu in mus]
    # T√≠nh to√°n r·ªßi ro v√† l·ª£i nhu·∫≠n t∆∞∆°ng ·ª©ng
    returns = [blas.dot(pbar, x) for x in portfolios]
    risks = [np.sqrt(blas.dot(x, S * x)) for x in portfolios]
    # T√¨m ƒëi·ªÉm MVP tr√™n ƒë∆∞·ªùng bi√™n hi·ªáu qu·∫£
    m1 = np.polyfit(returns, risks, 2)
    x1 = np.sqrt(m1[2] / m1[0])
    # Gi·∫£i b√†i to√°n t·ªëi ∆∞u l·∫ßn cu·ªëi ƒë·ªÉ l·∫•y tr·ªçng s·ªë MVP
    wt = solvers.qp(opt.matrix(x1 * S), -pbar, G, h, A, b)['x']
    return list(wt)  # tr·∫£ v·ªÅ danh s√°ch tr·ªçng s·ªë
# H√†m t√≠nh Inverse Variance Portfolio (IVP):
def getIVP(cov, **kargs):
    # T√≠nh danh m·ª•c t·ª∑ l·ªá ngh·ªãch v·ªõi ph∆∞∆°ng sai (kh√¥ng c·∫ßn t·ªëi ∆∞u h√≥a)
    ivp = 1. / np.diag(cov)  # 1/variance
    ivp /= ivp.sum()  # chu·∫©n h√≥a ƒë·ªÉ t·ªïng tr·ªçng s·ªë = 1
    return ivp
# H√†m t√≠nh Hierarchical Risk Parity (HRP):
def getHRP(cov, corr):
    # X√¢y d·ª±ng danh m·ª•c ph√¢n b·ªï r·ªßi ro theo ph√¢n c·ª•m ph√¢n c·∫•p (HRP) 
    dist = correlDist(corr)  # chuy·ªÉn ma tr·∫≠n t∆∞∆°ng quan th√†nh ma tr·∫≠n kho·∫£ng c√°ch
    link = sch.linkage(dist, 'single')  # ph√¢n c·ª•m theo ph∆∞∆°ng ph√°p ƒë∆°n li√™n k·∫øt (single linkage)
    # s·∫Øp x·∫øp l·∫°i theo th·ª© t·ª± b√°n ph√¢n c·ª•m (quasi-diagonalization)
    sortIx = getQuasiDiag(link)
    sortIx = corr.index[sortIx].tolist()
    # t√≠nh tr·ªçng s·ªë theo ph∆∞∆°ng ph√°p ph√¢n t√°ch ƒë·ªá quy
    hrp = getRecBipart(cov, sortIx)
    return hrp.sort_index()  # tr·∫£ v·ªÅ tr·ªçng s·ªë HRP theo th·ª© t·ª± ch·ªâ s·ªë ban ƒë·∫ßu
```

#### T√≠nh to√°n tr·ªçng s·ªë danh m·ª•c theo c√°c ph∆∞∆°ng ph√°p ph√¢n b·ªï t√†i s·∫£n

```python
def get_all_portfolios(returns):
    # T√≠nh ma tr·∫≠n hi·ªáp ph∆∞∆°ng sai v√† ma tr·∫≠n t∆∞∆°ng quan t·ª´ d·ªØ li·ªáu l·ª£i su·∫•t
    cov, corr = returns.cov(), returns.corr()
    # T√≠nh tr·ªçng s·ªë danh m·ª•c theo ph∆∞∆°ng ph√°p HRP (Hierarchical Risk Parity)
    hrp = getHRP(cov, corr)
    # T√≠nh tr·ªçng s·ªë danh m·ª•c theo ph∆∞∆°ng ph√°p MVP (Minimum Variance Portfolio)
    mvp = getMVP(cov)
    mvp = pd.Series(mvp, index=cov.index)
    # T·∫°o b·∫£ng t·ªïng h·ª£p c√°c tr·ªçng s·ªë t·ª´ hai ph∆∞∆°ng ph√°p MVP v√† HRP
    portfolios = pd.DataFrame([mvp, hrp], index=['MVP', 'HRP']).T
    # N·∫øu mu·ªën thay ƒë·ªïi sang IVP, c√≥ th·ªÉ thay d√≤ng tr√™n b·∫±ng d√≤ng d∆∞·ªõi (ƒë√£ ƒë∆∞·ª£c comment)
    # portfolios = pd.DataFrame([ivp, hrp], index=['IVP', 'HRP']).T
    
    return portfolios
```

```python
portfolios = get_all_portfolios(returns)
```

M·∫∑t kh√°c, IVP ph√¢n b·ªï tr·ªçng s·ªë ƒë·ªÅu cho t·∫•t c·∫£ c√°c t√†i s·∫£n, v√† n√≥ tr√¥ng kh√° gi·ªëng v·ªõi HRP. B√¢y gi·ªù, h√£y c√πng xem k·∫øt qu·∫£ th·ª≠ nghi·ªám h·ªìi quy (backtesting).

```python
Insample_Result = pd.DataFrame(np.dot(returns, np.array(portfolios)), \
                              columns=['MVP', 'HRP'], index=returns.index)
# T·∫°o DataFrame k·∫øt qu·∫£ th·ª≠ nghi·ªám tr√™n d·ªØ li·ªáu trong m·∫´u (Insample), v·ªõi c·ªôt l√† c√°c ph∆∞∆°ng ph√°p MVP v√† HRP, ch·ªâ s·ªë l√† th·ªùi gian c·ªßa returns.
OutOfSample_Result = pd.DataFrame(np.dot(returns_test, np.array(portfolios)), \
                                 columns=['MVP', 'HRP'], index=returns_test.index)
# T·∫°o DataFrame k·∫øt qu·∫£ th·ª≠ nghi·ªám tr√™n d·ªØ li·ªáu ngo√†i m·∫´u (OutOfSample), v·ªõi c·ªôt l√† c√°c ph∆∞∆°ng ph√°p MVP v√† HRP, ch·ªâ s·ªë l√† th·ªùi gian c·ªßa returns_test.
```

### K·∫øt qu·∫£ trong m·∫´u v√† ngo√†i m·∫´u

```python
# K·∫øt qu·∫£ trong m·∫´u (In-sample Results)
stddev = Insample_Result.std() * np.sqrt(252)  # T√≠nh ƒë·ªô l·ªách chu·∫©n h√†ng nƒÉm (r·ªßi ro)
sharp_ratio = (Insample_Result.mean() * np.sqrt(252)) / (Insample_Result.std())  # T√≠nh t·ª∑ s·ªë Sharpe h√†ng nƒÉm
Results = pd.DataFrame(dict(stdev=stddev, sharp_ratio=sharp_ratio))  # T·∫°o b·∫£ng k·∫øt qu·∫£ v·ªõi ƒë·ªô l·ªách chu·∫©n v√† t·ª∑ s·ªë Sharpe
Results
```
